5.What is mutability in Python ?
 - Mutability in Python refers to the ability of an object to be modified after it is created. If an object is mutable, its state or contents can be changed without creating a new object. Conversely, if an object is immutable, its value is fixed once it's created, and any operation that appears to modify it will actually result in the creation of a new object.
Common mutable data types in Python include:
Lists: Elements can be added, removed, or modified.
Dictionaries: Key-value pairs can be added, removed, or updated.
Sets: Elements can be added or removed.

6.Why are lists mutable, but tuples are immutable
 - In Python, lists are considered mutable because you can change their elements after they are created, while tuples are immutable, meaning once a tuple is created, you cannot modify its elements; essentially, lists are designed for data that needs to be frequently updated, while tuples are used to store data that should remain unchanged, ensuring data integrity and sometimes providing slight performance benefits due to their immutability.

7.What is the difference between “==” and “is” operators in Python ?
 - The == and is operators in Python serve distinct comparison purposes. The == operator checks for equality of value between two objects, while the is operator checks for identity, verifying if two variables refer to the exact same object in memory.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print(list1 == list2) # Output: True, because the lists have the same values
print(list1 is list2) # Output: False, because they are different objects in memory
print(list1 is list3) # Output: True, because list3 refers to the same object as list1
```


8.What are logical operators in Python ?

9.What is type casting in Python ?
 - Type casting in Python, also known as type conversion, is the process of changing a variable's data type into another data type. It's essential for ensuring compatibility between different data types and enhancing the versatility of Python code. Type casting can be broadly categorized into two types: implicit (automatic) casting and explicit (manual) casting.

10.What is the difference between implicit and explicit type casting ?
 -  Implicit type casting happens automatically, while explicit type casting requires the programmer to intervene.

∆ Implicit type casting

• Also known as widening conversion

• Automatically converts a smaller data type to a larger data type

• Ensures compatibility and prevents data loss in mathematical operations

• Often invisible to the user

∆ Explicit type casting

• Also known as narrowing conversion

• Manually converts a larger data type to a smaller data type

• Provides more control over the data conversion process

• Visible in the source code

• Prevents unintended data loss or errors

11.What is the purpose of conditional statements in Python ?
 - The purpose of conditional statements in Python is to control the flow of execution in a program based on whether certain conditions are met. Conditional statements allow a program to make decisions, executing different blocks of code depending on the outcome of a condition. The primary conditional statements in Python are if, elif (else if), and else.

12.How does the elif statement work ?
 - In Python, "elif" stands for "else if" and is used to check additional conditions if the first "if" statement is false; essentially, it allows you to evaluate multiple conditions sequentially, executing the block of code only if the current "elif" condition is true, while skipping to the next "elif" if the previous one is false.

13.What is the difference between for and while loops ?
 - Both for loop and while loop is used to execute the statements repeatedly while the program runs. The major difference between for loop and while loop is that for loop is used when the number of iterations is known, whereas execution is done in a while loop until the statement in the program is proved wrong.

14.Describe a scenario where a while loop is more suitable than a for loop.
 - In contrast to a 'for' loop, which is generally used when the number of iterations is known beforehand, a 'while' loop is ideal for situations where the number of iterations is not predetermined and depends on some dynamic condition.
